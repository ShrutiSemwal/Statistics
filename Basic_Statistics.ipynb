```python
DF = pd.read_csv('Housing.csv')
```

```python
DF
```

```python
print("Mean of DataFrame['area'] : ", DF['area'].mean())
```

```python
print("Median of DataFrame['area'] : ", DF['area'].median())
```

```python
print("Mode of DataFrame['area'] : ", DF['area'].mode())
```

```python
print("Standard Deviation of DataFrame['area'] : ", DF['area'].std())
```

# **Visualisation**

```python
DF['area'].plot()
```

```python
sns.scatterplot(data=DF, x="area", y="price", hue='bedrooms', size='parking')
```

```python
sns.scatterplot(data=DF, x="area", y="price")
```

```python
sns.scatterplot(data=DF, x="area", y="price", hue='bedrooms', size='parking', palette=sns.color_palette("rocket", as_cmap=False))
```

```python
sns.lineplot(data=DF, x="area", y="price")
```

```python
sns.lineplot(data=DF, x="bedrooms", y="bathrooms", hue='stories', palette=sns.color_palette("YlOrBr", as_cmap=True))
```

```python
sns.histplot(data=DF, x="bedrooms", y="bathrooms", hue='stories', palette=sns.color_palette("YlOrBr", as_cmap=True))
```

```python
sns.histplot(data=DF, x="bedrooms", y="bathrooms", hue='stories', palette=sns.color_palette("Spectral", as_cmap=False))
```

```python
sns.displot(data=DF, x="bedrooms", y="bathrooms", hue='stories', palette=sns.color_palette("Spectral", as_cmap=False))
```

```python
sns.boxplot(data=DF, x="bedrooms", y="bathrooms", hue='stories', palette=sns.color_palette("Spectral", as_cmap=False))
```
